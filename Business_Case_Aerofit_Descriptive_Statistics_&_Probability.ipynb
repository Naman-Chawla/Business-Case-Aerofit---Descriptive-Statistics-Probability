### Conditional probability of Product w.r.t gender and vice versa



```python
# Count of each Product w.r.t Gender
PGC = pd.crosstab(df["Gender"], df["Product"], margins = True)

# P(Product/Gender)
PG = pd.crosstab(df["Gender"], df["Product"], margins = True, normalize = "index")

# P(Gender/Product)
GP = pd.crosstab(df["Gender"], df["Product"], margins = True, normalize = "columns")
print(f"Count of each Product w.r.t Gender:\n{PGC} \n\nP(Product/Gender): \n{PG} \n\nP(Gender/Product): \n{GP}")
```

Obser:\
Female more likely to buy KP281 (irrespective of other factors)\
There is no major difference between Prob. of Males buying any of the Tredmill.\
Prob. of being a male given that customer had bought KP781 is highest, its around 0.825

### Conditional probability of Product w.r.t MaritalStatus and vice versa

```python
# Count of each Product w.r.t MaritalStatus
PMC = pd.crosstab(df["MaritalStatus"], df["Product"], margins = True)

# P(Product/MaritalStatus)
PM = pd.crosstab(df["MaritalStatus"], df["Product"], margins = True, normalize = "index")

# P(MaritalStatus/Product)
MP = pd.crosstab(df["MaritalStatus"], df["Product"], margins = True, normalize = "columns")
print(f"Count of each Product w.r.t MaritalStatus:\n{PMC} \n\nP(Product/MaritalStatus): \n{PM} \n\nP(MaritalStatus/Product): \n{MP}")
```

Obser:\
Prob. of buying KP281 in both the cases (customer being single or partnered) is\
almost same and maxinmum among other Prodcuts.

### Conditional probability of Product w.r.t (Fitness and Gender)

```python
# Data is filtered according to two cond. (fitness = 3 i.e customers having Average fitness and Gender == Male or Female)
FnG = df[df["Fitness"] == 3] # Customers with Average Fitness

FnGC = pd.crosstab(FnG["Gender"], FnG["Product"], margins = True)

# P(Product/Fitness and Gender)
FGP = pd.crosstab(FnG["Gender"], FnG["Product"], margins = True, normalize = "index")
print(f"Count of each Product w.r.t Fitness and Gender:\n{FnGC} \n\nP(Product/Fitness and Gender): \n{FGP}")
```
